```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# Carregando o dataset
df = pd.read_csv("obesidade.csv")
```

```python

# Exibindo as primeiras linhas do dataset
print("Primeiras linhas do dataset:")
print(df.head())
```

```python
# Exibindo informações sobre o dataset
print("\nInformações sobre o dataset:")
df.info()
```

```python
# Exibindo a quantidade de valores nulos
print("\nQuantidade de valores nulos por coluna:")
print(df.isnull().sum())
```

```python
# Exibindo a quantidade de valores únicos
print("\nQuantidade de valores únicos por coluna:")
print(df.nunique())
```

```python
# Exibindo a descrição estatística do dataset
print("\nDescrição estatística do dataset:")
print(df.describe())
```

```python
# Tratamento de valores nulos
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Normalizando as colunas numéricas
numeric_df = df.select_dtypes(include=[np.number])
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(
    scaler.fit_transform(numeric_df), columns=numeric_df.columns
    )
```

```python
# Plotando a distribuição das variáveis numéricas normalizadas
print("\nDistribuição das variáveis numéricas normalizadas:")
normalized_df.hist(bins=30, figsize=(20, 15))
plt.show()
```
